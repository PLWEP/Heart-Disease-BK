```python
import pandas as pd
```

```python
file_path = 'heart+disease\hungarian.data'

with open(file_path, 'r') as file:
    raw_data = file.read()

raw_data = raw_data.strip().split('name')
raw_data.pop()
cleaned_data = [string.replace('\n\n', ' ').replace('\n', ' ').strip().split(' ') for string in raw_data]
data = pd.DataFrame(cleaned_data)
column_feature = [2, 3, 8, 9, 11, 15, 18, 31, 37, 39, 40, 43, 50, 57]
data_filtered = data.iloc[:, column_feature]

# Dictionary untuk pemetaan nama kolom baru
column_mapping = {
    2: 'age',
    3: 'sex',
    8: 'cp',
    9: 'trestbps',
    11: 'chol',
    15: 'fbs',
    18: 'restecg',
    31: 'thalach',
    37: 'exang',
    39: 'oldpeak',
    40: 'slope',
    43: 'ca',
    50: 'thal',
    57: 'num',
}

# Mengganti nama kolom sesuai dengan pemetaan
data_filtered = data_filtered.rename(columns=column_mapping)

data_filtered
```

```python
file_path = r'heart+disease\reprocessed.hungarian.data'

with open(file_path, 'r') as file:
    raw_data = file.read()

raw_data = raw_data.strip().split('\n')
cleaned_data = [string.strip().split(' ') for string in raw_data]
data = pd.DataFrame(cleaned_data, columns=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num'])

data
```
